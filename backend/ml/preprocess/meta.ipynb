```python
import datasets
from datasets import load_dataset
```

```python
dataset = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "raw_meta_Electronics", split="full", trust_remote_code=True)
```

```python
dataset[0]
```

```python
len(dataset)
```

```python
from tqdm import tqdm
meta_id, meta_item_id = [], []
for idx, data in enumerate(tqdm(dataset)):
    meta_id.append(idx)
    meta_item_id.append(data["parent_asin"])
```

```python
import pandas as pd
meta_df = pd.DataFrame({
    "meta_id": meta_id,
    "meta_item_id": meta_item_id
})
meta_df
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/raw/Electronics.csv")
df
```

```python
unique_item_id = df["parent_asin"].unique()
len(unique_item_id)
```

```python
unique_item_df = pd.DataFrame({
    "meta_item_id": unique_item_id
})
unique_item_df
```

```python
item_df = pd.merge(meta_df, unique_item_df, how="inner", on="meta_item_id")
item_df
```

```python
item_df.to_pickle("../../data/interim/meta.pkl")
```

```python
meta_json = [dataset[idx] for idx in item_df["meta_id"].tolist()]
import json
json.dump(meta_json, open("../../data/interim/meta.json", "w"))
```

```python
df = pd.DataFrame({
    "title": [item["title"] for item in meta_json], 
    "categories":[item["categories"] for item in meta_json], 
    "main_category":[item["main_category"] for item in meta_json], 
    "average_rating": [item["average_rating"] for item in meta_json], 
    "description": [item["description"] for item in meta_json],  
    "images": [item["images"] for item in meta_json], 
    "price":[item["price"] for item in meta_json], 
    "parent_asin": [item["parent_asin"] for item in meta_json], 
    "rating_number": [item["rating_number"] for item in meta_json], 
})
df
```
