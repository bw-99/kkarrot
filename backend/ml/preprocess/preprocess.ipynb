```python
import pandas as pd
import numpy as np
NUM_USER, NUM_ITEM = 2_000, 300_000
df = pd.read_csv("../../data/raw/Electronics.csv")
df
```

```python
top_user_df = df.groupby("user_id").agg({
    "parent_asin": lambda x: x.tolist(),
    "rating": lambda x: x.tolist(),
    "timestamp": lambda x: x.tolist(),
}).reset_index(drop=False)
top_user_df["item_len"] = top_user_df["rating"].apply(lambda x: len(x))
top_user_df
```

```python
item_lens = top_user_df["item_len"].to_numpy()

top_user_df = top_user_df.iloc[np.argsort(item_lens)[::-1][:NUM_USER]].reset_index(drop=True)
top_user_df
```

```python
top_item_df = df.groupby("parent_asin").agg({
    "user_id": lambda x: x.tolist(),
    "rating": lambda x: x.tolist(),
    "timestamp": lambda x: x.tolist(),
}).reset_index(drop=False)
top_item_df["user_len"] = top_item_df["rating"].apply(lambda x: len(x))
top_item_df
```

```python
user_lens = top_item_df["user_len"].to_numpy()

top_item_df = top_item_df.iloc[np.argsort(user_lens)[::-1][:NUM_ITEM]].reset_index(drop=True)
top_item_df
```

```python
top_user_df = top_user_df.explode(["parent_asin", "rating", "timestamp"])
```

```python
top_item_df = top_item_df.explode(["user_id", "rating", "timestamp"])
```

```python
ddf = pd.merge(top_item_df, top_user_df, on=["user_id", "parent_asin"], how="inner")[["parent_asin", "user_id", "timestamp_x", "rating_x"]]
ddf = ddf.rename(columns={
    "rating_x": "rating", 
    "timestamp_x": "timestamp"
})
ddf
```

```python
ddf = ddf.groupby("parent_asin").agg({
    "user_id": lambda x: x.tolist(),
    "rating": lambda x: x.tolist(),
    "timestamp": lambda x: x.tolist(),
}).reset_index(drop=False)

ddf["item_id"] = [i for i in range(len(ddf))]
ddf = ddf.explode(["user_id", "rating", "timestamp"])
ddf
```

```python
ddf = ddf.groupby("user_id").agg({
    "item_id": lambda x: x.tolist(),
    "parent_asin": lambda x: x.tolist(),
    "rating": lambda x: x.tolist(),
    "timestamp": lambda x: x.tolist(),
}).reset_index(drop=False)

ddf["user_id"] = [i for i in range(len(ddf))]
ddf = ddf.explode(["item_id", "parent_asin", "rating", "timestamp"])
ddf
```
